```python
import os
import concurrent.futures
from dashscope import MultiModalConversation
from typing import List, Dict
from datetime import datetime

def process_single_image(image_path: str, question: str, api_key: str) -> str:
    """处理单张图片并返回结果"""
    try:
        local_path = os.path.abspath(image_path)
        image_url = f"file://{local_path}"
        
        messages = [
            {"role": "system", "content": [{"text": "You are a helpful assistant."}]},
            {'role': 'user', 'content': [
                {'image': image_url},
                {'text': question}
            ]}
        ]
        
        response = MultiModalConversation.call(
            api_key=api_key,
            model='qwen2.5-vl-7b-instruct', 
            messages=messages
        )
        
        return response["output"]["choices"][0]["message"].content[0]["text"]
    
    except Exception as e:
        return f"处理图片 {image_path} 时出错: {str(e)}"

def batch_process_images(image_paths: List[str], question: str, api_key: str, max_workers: int = 5) -> Dict[str, str]:
    """批量处理多张图片"""
    results = {}
    
    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        # 创建任务映射
        future_to_path = {
            executor.submit(process_single_image, path, question, api_key): path 
            for path in image_paths
        }
        
        # 等待所有任务完成并收集结果
        for future in concurrent.futures.as_completed(future_to_path):
            image_path = future_to_path[future]
            try:
                results[image_path] = future.result()
                print(f"已处理图片: {os.path.basename(image_path)}")
            except Exception as e:
                results[image_path] = f"处理失败: {str(e)}"
    
    return results

start_time = datetime.now()
# 使用示例
# if __name__ == "__main__":
api_key = "sk-e4f5d5efbab54ebd9207f3865426edfd"

# 批量图片路径
image_paths = [
    "D:\\ultralytics\\YOLO+CLIP\\YOLO+CLIP_output_T2I_I2I\\636-0.9595-0.9936-0.9414-crop.jpg",
    "D:\\ultralytics\\YOLO+CLIP\\YOLO+CLIP_output_T2I_I2I\\234-0.8175-0.6119-0.8871-crop.jpg",
    "D:\\ultralytics\\YOLO+CLIP\\YOLO+CLIP_output_T2I_I2I\\1374-0.9172-0.9106-0.9208-crop.jpg",
    "D:\\ultralytics\\YOLO+CLIP\\YOLO+CLIP_output_T2I_I2I\\1494-0.8458-0.9554-0.7879-crop.jpg",
    "D:\\ultralytics\\YOLO+CLIP\\YOLO+CLIP_output_T2I_I2I\\1044-0.8367-0.9991-0.7817-crop.jpg",
    # 添加更多图片路径...
]

question = "图中描绘的是什么景象?"

# 批量处理
results = batch_process_images(image_paths, question, api_key, max_workers=5)

# 输出结果
for image_path, description in results.items():
    print(f"图片: {os.path.basename(image_path)}")
    print(f"描述: {description}")
    print("-" * 50)

end_time = datetime.now()
duration = end_time - start_time
print(f"处理完成！共处理了 {len(results)} 张图片，耗时: {duration}")
```

### 使用异步处理

```python
import asyncio
import aiohttp
import os
from typing import List, Dict

async def async_process_image(session, image_path: str, question: str, api_key: str) -> str:
    """异步处理单张图片"""
    try:
        local_path = os.path.abspath(image_path)
        image_url = f"file://{local_path}"
        
        # 这里需要根据 dashscope 的实际异步API进行调整
        # 以下为示例代码，可能需要使用 dashscope 的异步客户端
        payload = {
            "model": "qwen2.5-vl-7b-instruct",
            "messages": [
                {"role": "system", "content": [{"text": "You are a helpful assistant."}]},
                {"role": "user", "content": [
                    {"image": image_url},
                    {"text": question}
                ]}
            ]
        }
        
        headers = {
            "Authorization": f"Bearer {api_key}",
            "Content-Type": "application/json"
        }
        
        async with session.post(
            "https://dashscope.aliyuncs.com/api/v1/services/aigc/multimodal-generation/generation",
            json=payload,
            headers=headers
        ) as response:
            result = await response.json()
            return result["output"]["choices"][0]["message"].content[0]["text"]
            
    except Exception as e:
        return f"处理图片 {image_path} 时出错: {str(e)}"

async def async_batch_process(image_paths: List[str], question: str, api_key: str) -> Dict[str, str]:
    """异步批量处理"""
    results = {}
    
    async with aiohttp.ClientSession() as session:
        tasks = []
        for path in image_paths:
            task = async_process_image(session, path, question, api_key)
            tasks.append(task)
        
        # 并行执行所有任务
        descriptions = await asyncio.gather(*tasks, return_exceptions=True)
        
        for path, description in zip(image_paths, descriptions):
            if isinstance(description, Exception):
                results[path] = f"处理失败: {str(description)}"
            else:
                results[path] = description
    
    return results

# 使用示例
async def main():
    api_key = "sk-e4f5d5efbab54ebd9207f3865426edfd"
    image_paths = [
        "D:\\ultralytics\\YOLO+CLIP\\YOLO+CLIP_output_T2I_I2I\\636-0.9595-0.9936-0.9414-crop.jpg",
        "D:\\ultralytics\\YOLO+CLIP\\YOLO+CLIP_output_T2I_I2I\\234-0.8175-0.6119-0.8871-crop.jpg",
    ]
    
    question = "图中描绘的是什么景象?"
    
    results = await async_batch_process(image_paths, question, api_key)
    
    for image_path, description in results.items():
        print(f"图片: {os.path.basename(image_path)}")
        print(f"描述: {description}")
        print("-" * 50)

# 运行异步主函数
if __name__ == "__main__":
    asyncio.run(main())
```

### 目录批量处理

```python
import os
import concurrent.futures
from dashscope import MultiModalConversation
from typing import List, Dict

def process_directory_images(directory_path: str, question: str, api_key: str, 
                           max_workers: int = 5, image_extensions: List[str] = None) -> Dict[str, str]:
    """处理目录中的所有图片"""
    if image_extensions is None:
        image_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff']
    
    # 获取目录中所有图片文件
    image_paths = []
    for filename in os.listdir(directory_path):
        if any(filename.lower().endswith(ext) for ext in image_extensions):
            image_paths.append(os.path.join(directory_path, filename))
    
    return batch_process_images(image_paths, question, api_key, max_workers)

# 使用示例
if __name__ == "__main__":
    api_key = "sk-e4f5d5efbab54ebd9207f3865426edfd"
    image_directory = "D:\\ultralytics\\YOLO+CLIP\\YOLO+CLIP_output_T2I_I2I"
    question = "图中描绘的是什么景象?"
    
    # 处理整个目录
    results = process_directory_images(image_directory, question, api_key, max_workers=4)
    
    # 保存结果到文件
    with open("image_descriptions.txt", "w", encoding="utf-8") as f:
        for image_path, description in results.items():
            f.write(f"图片: {os.path.basename(image_path)}\n")
            f.write(f"描述: {description}\n")
            f.write("-" * 50 + "\n")
    
    print(f"处理完成！共处理了 {len(results)} 张图片，结果已保存到 image_descriptions.txt")
```
